# **Import necessary libraries**

```python
import pandas as pd
import numpy as np
import yfinance as yf
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
```

# **Download historical stock data for Tesla**
# use the ticker symbol 'TSLA' for Tesla

```python
tesla_data = yf.download('TSLA', start='2018-01-01', end='2023-01-01')
```

## Display the first few rows of the data

```python
print("First few rows of Tesla data:")
print(tesla_data.head())
```

## **Data Preprocessing and Feature Engineering**
We add some useful features to make our predictions more accurate

```python
tesla_data['Moving_Avg_10'] = tesla_data['Close'].rolling(window=10).mean()
tesla_data['Moving_Avg_50'] = tesla_data['Close'].rolling(window=50).mean()
tesla_data = tesla_data.dropna()  # Drop rows with missing values
```

# Display first few rows after adding new features

```python
print("\nData with Moving Averages:")
print(tesla_data.head())
```

# **Selecting Features and Target Variable**
# Features: 'Moving_Avg_10' and 'Moving_Avg_50' (moving averages of close prices)
# Target: 'Close' (closing price of the stock)

```python
X = tesla_data[['Moving_Avg_10', 'Moving_Avg_50']]
y = tesla_data['Close']
```

#  Splitting Data into Training and Testing Sets
#  split our data into 80% training and 20% testing

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

# Train a Linear Regression Model


```python
model = LinearRegression()
model.fit(X_train, y_train)
```

Making Predictions

```python
y_pred = model.predict(X_test)
```

Model Evaluation

```python
mse = mean_squared_error(y_test, y_pred)
print(f"\nMean Squared Error (MSE): {mse}")
```

Visualization of Predicted vs Actual Prices

```python
plt.figure(figsize=(10, 5))
plt.plot(y_test.index, y_test, label='Actual Prices', color='blue')
plt.plot(y_test.index, y_pred, label='Predicted Prices', color='red')
plt.xlabel("Date")
plt.ylabel("Stock Price (Close)")
plt.title("Tesla Stock Price Prediction")
plt.legend()
plt.show()
```
